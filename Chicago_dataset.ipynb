> Data Preprocessing: the column "Trip Start Timestamp" should be divided into 6 others:\
Minute, Hour, Day, Month, Day of week, Week of year. These parameters make sense when we build models related to regression, for example. Because feature engineering prove crucial to achieve higher accuracy.

```python
df["Trip Start Timestamp"] = pd.to_datetime(df["Trip Start Timestamp"], format='%m/%d/%Y %I:%M:%S %p')

df["Minute"] = df["Trip Start Timestamp"].dt.minute
df["Hour"] = df["Trip Start Timestamp"].dt.hour
df["Day"] = df["Trip Start Timestamp"].dt.day
df["Month"] = df["Trip Start Timestamp"].dt.month
df['weekday'] = df['Trip Start Timestamp'].dt.weekday_name
df['week_year'] = df['Trip Start Timestamp'].dt.weekofyear
```

> Data Preprocessing: I also changed string values into numeric representation of them

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
df['Payment Type_num'] = le.fit_transform(df['Payment Type'])
df['Company_num'] = le.fit_transform(df['Company'])
df['weekday_num'] = le.fit_transform(df['weekday'])
```

> Choosing two columns from this data (Longitude and Latitude) is the way of data preprocessing

```python
X = df[['Pickup Centroid Longitude','Pickup Centroid Latitude']]
X = X.to_numpy() #we need array
X
```

```python
from sklearn.cluster import KMeans 
from sklearn import metrics 
from scipy.spatial.distance import cdist 
import numpy as np 
import matplotlib.pyplot as pl
distortions = [] 
inertias = [] 
mapping1 = {} 
mapping2 = {} 
K = range(1,10) 
  
for k in K: 
    #Building and fitting the model 
    kmeanModel = KMeans(n_clusters=k).fit(X) 
    kmeanModel.fit(X)     
      
    distortions.append(sum(np.min(cdist(X, kmeanModel.cluster_centers_, 
                      'euclidean'),axis=1)) / X.shape[0]) 
    inertias.append(kmeanModel.inertia_) 
  
    mapping1[k] = sum(np.min(cdist(X, kmeanModel.cluster_centers_, 
                 'euclidean'),axis=1)) / X.shape[0] 
    mapping2[k] = kmeanModel.inertia_ 
```

```python
plt.plot(K, distortions, 'bx-') 
plt.xlabel('Values of K') 
plt.ylabel('Distortion') 
plt.title('The Elbow Method using Distortion') 
plt.show() 
```

> Graph interpretation: to determine the optimal number of clusters, we have to select the value of k at the “elbow” i.e. the point after which the distortion start decreasing in a linear trajectory. Thus for the given data, we conclude that the optimal number of clusters for the data is 3 (rather than 2)

>Remark: we will run both number of clusters (2 and 3) to assess the performance

> Method: **KMeans**:\
I used this method for this dataset because it scales to large data sets with not a huge number of features. Here, we see that the number of clusters is equal to 2 or 3. K-Means guarantees converenge (because geometry metric). Also, in K-Means we use distances between all points like the metric.  

```python
labels_2 = KMeans(2, random_state=0).fit_predict(X)
kmeans_2 = KMeans(n_clusters=2, random_state=0).fit(X)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Chicago Trips k = 2', size =15)
plt.scatter(X[:, 0], X[:, 1], c=labels_2,
            s=20, cmap='viridis')
plt.scatter(kmeans_2.cluster_centers_[0][0],kmeans_2.cluster_centers_[0][1], color = 'red', s = 100)
plt.scatter(kmeans_2.cluster_centers_[1][0],kmeans_2.cluster_centers_[1][1], color = 'chocolate', s = 100)
```

> Graph interpretations: In the case with 2 clusters, we see something strange because K Means has divided into two clusters with unequal sizes. However, it looks like almost the whole of Chicago is in the cluster with purple dots and the red centroid. It makes sense if we try to distinguish urban taxi trips with more rural (yellow points with chocolate centroid).

```python
labels_3 = KMeans(3, random_state=0).fit_predict(X)
kmeans_3 = KMeans(n_clusters=3, random_state=0).fit(X)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Chicago Trips k = 3', size =15)
plt.scatter(X[:, 0], X[:, 1], c=labels_3,
            s=20, cmap='viridis');
plt.scatter(kmeans_3.cluster_centers_[0][0],kmeans_3.cluster_centers_[0][1], color = 'red', s = 100)
plt.scatter(kmeans_3.cluster_centers_[1][0],kmeans_3.cluster_centers_[1][1], color = 'chocolate', s = 100)
plt.scatter(kmeans_3.cluster_centers_[2][0],kmeans_3.cluster_centers_[2][1], color = 'pink', s =100)
```

> Graph interpretations: initially, the graph with 3 clusters separate Chicago taxi trips into the north and the south parts of the city (yellow with pink centroid and purple with red centroid respectively). It may have essential information because we know that there are a lot of sights in the north of Chicago. Hence, people will use more taxis there, especially tourists, as compared to the south part, which is focused on the financial district and everything related to that. People there may also frequently use a taxi as in the north, but the trips in the financial district may be shorter than in the north part of the city. The third cluster with chocolate centroid is connected with more rural locations. I **conclude** that 3 clusters can be more beneficial for further analysis. Especially, we saw from the Elbow graph that either 2 or 3 makes sense for us. In terms of metrics (in clusterisation we do not know the "true" labels), I want to focus on the location of centroids. The centroid in the purple cluster seems like a mistake because the distances are not stable from centroid to each point from its cluster. But the performance of the algorithms depends on initial position of centroids and distribution of data.
